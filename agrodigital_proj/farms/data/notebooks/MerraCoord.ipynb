```python
# Open the NetCDF file
with nc.Dataset(file_path, 'r') as netcdf_file:
    # Print basic information about the NetCDF file
    print("NetCDF file information:")
    print("Dimensions:")
    for dimname, dimobj in netcdf_file.dimensions.items():
        print(f"  {dimname}: {len(dimobj)}")

    print("\nVariables:")
    for varname, varobj in netcdf_file.variables.items():
        print(f"  {varname}: {varobj.shape} {varobj.units if 'units' in varobj.ncattrs() else ''}")
```
